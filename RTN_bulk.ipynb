# Transcriptional Network Analysis (TNA)

```r
list.files('/tank/projects/diana_omics/bulk_rna/DGE_DESeq2/')
```

```r
load('new_tni_filt')
```

```r
dif <- read.csv('/tank/projects/diana_omics/bulk_rna/DGE_DESeq2/sig_genes.csv', sep=';')
```

```r
dif$gene_name <- mapIds(org.Hs.eg.db, dif$gene, column = "SYMBOL", "ENSEMBL")
```

```r
require(dplyr)

dif<- dif %>% relocate(gene_name, .after = gene)
```

```r
dif = subset(dif, select = -c(gene) )
dif <- na.omit(dif)
head(dif)
```

```r
pheno <- setNames(dif$log2FoldChange, dif$gene_name)
```

```r
rtna <- tni2tna.preprocess(object = rtni, 
                           phenotype = pheno, 
                           hits = dif$gene_name)
```

```r
rtna <- tna.mra(rtna)
```

```r
mra <- tna.get(rtna, what="mra", ntop = -1)
head(mra)
```

```r
# Run the GSEA method
# Please set nPermutations >= 1000
rtna <- tna.gsea1(rtna, nPermutations=1000)
```

```r
# Get GSEA results
gsea1 <- tna.get(rtna, what="gsea1", ntop = -1)
head(gsea1)
```

```r
write.csv(gsea1, "gsea1.csv")
```

```r
tna.plot.gsea1(rtna, labPheno="abs(log2 fold changes)", ntop = -1)
```

```r
# Run the GSEA-2T method
# Please set nPermutations >= 1000
rtna <- tna.gsea2(rtna, nPermutations = 1000)
```

```r
gsea2 <- tna.get(rtna, what = "gsea2", ntop = -1)
head(gsea2$differential)
```

```r
write.csv(gsea2, "gsea2.csv")
```

```r
tna.plot.gsea2(rtna, labPheno="log2 fold changes", tfs="ZNF628")
```
